# 4. GDP-VAR

```python
gdp_raw = pd.read_csv('/content/drive/My Drive/Final Project Boolean/Final_Raw_Data/01_Dataset_final-project - GDP-VAR.csv')
gdp = pd.DataFrame(gdp_raw)
gdp.head()
```

```python
# Seleziona solo le colonne degli anni (1980-2028)
colonne_anni = [str(anno) for anno in range(1980, 2029)]

# Converti le colonne degli anni in numeriche
gdp[colonne_anni] = gdp[colonne_anni].apply(pd.to_numeric, errors='coerce')

print(gdp.dtypes)
```

```python
# Seleziona solo le colonne degli anni (1980-2028)
colonne_anni = [str(anno) for anno in range(1980, 2029)]

# Converti le colonne degli anni in numeriche
gdp[colonne_anni] = gdp[colonne_anni].apply(pd.to_numeric, errors='coerce')

print(gdp.dtypes)
```

```python
gdp.isnull().sum()
```

```python
media_per_regione = gdp.groupby('STRAT-3')['1980'].mean()
print(media_per_regione)
```

```python
missing_region = gdp[gdp['STRAT-3'].isnull()]
missing_region
```

```python
# Definisci una funzione per sostituire i valori nulli con la media della regione per un determinato anno
def fill_na_with_region_mean(series):
    return series.fillna(series.groupby(gdp['STRAT-3']).transform('mean'))

# Applica la funzione a ciascuna colonna degli anni (dal 1980 al 2028)
for anno in range(1980, 2028):
    colonna_anno = f'{anno}'
    gdp[colonna_anno] = fill_na_with_region_mean(gdp[colonna_anno])
```

```python
gdp.head()
```

```python
gdp.isnull().sum()
```

```python
gdp = gdp.dropna()
gdp.isnull().sum()
```

```python
gdp.to_csv('Clean_GDP.csv', index=False)
```
